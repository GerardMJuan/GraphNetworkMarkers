Small scripts that compute delay matrices from tract lengths and best
cs values.

From now, it only compares histograms of tract lengths matrices (from the averages tract 
length of each subject), and tries to infer a good cs ot have understandable delays.

```python
import numpy as np
import os
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# path of the averages
avg_path = "/mnt/Bessel/Gproj/Gerard_DATA/MAGNIMS2021/avg"
list_of_centers = ["CLINIC", "MAINZ", "MILAN", "NAPLES", "OSLO"]

for center in list_of_centers:
    print(center)
    Tract_path = f"{avg_path}/{center}/results/tract_lengths.txt"
    Tract = np.loadtxt(open(Tract_path, "rb"), delimiter=" ", skiprows=0)
    print("Average value: ", np.mean(Tract))
    print("Max value: ", np.max(Tract))

    # draw the histogram
    plt.figure()
    plt.hist(Tract)
    plt.show()
```
